# 2. Reshape & slice

```python
import torch
my_torch = torch.arange(10)
my_torch
```

```python
# Reshape & View
my_torch = my_torch.reshape(2,5)
my_torch
```

```python
# Reshape if we not sure the number of items using -1
my_torch2 = torch.arange(15)
my_torch2 = my_torch.reshape(3,-1) # even when the data in storage is not contiguous, it still works
my_torch2
```

```python
my_torch3 = torch.arange(10)
my_torch4 = my_torch3.reshape(2,5)
my_torch3[1] = 5783
my_torch4
```

```python
# Slices
my_torch5 = torch.arange(10)
my_torch5[7]
```

```python
my_torch5 = my_torch5.reshape(5,-1)
my_torch5
```

```python
my_torch5[:,0]
```

```python
# Return column
my_torch5[:,1:]
```

# 3. Tensor Math operation


```python
import torch
import numpy as np

tensorA = torch.tensor([1,2,3])
tensorB = torch.tensor([4,5,6])

# addition
tensorA + tensorB

# addition longhand 1
torch.add(tensorA,tensorB)

# addition longhand 2
tensorA.add(tensorB)

# local operation 1
tensorA = torch.add(tensorA,tensorB)

# local operation 2
tensorA.add_(tensorB)
```

```python
# subtraction
tensorB - tensorA

# subtraction longhand
torch.sub(tensorA,tensorB)
```

```python
# multiplication
torch.mul(tensorA,tensorB)
```

```python
# division
torch.div(tensorA,tensorB)
```

```python
# modulus
torch.remainder(tensorB,tensorA)
```

```python
# exponent
torch.pow(tensorA,tensorB)
```
